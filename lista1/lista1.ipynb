```python
def calc_clouds(K,d,N):

    alfa = 2*np.pi/K
    rotation_matrix = np.array([[np.cos(alfa),np.sin(alfa)],[-np.sin(alfa),np.cos(alfa)]])

    centres = np.zeros(2*K).reshape(K,2)
    points = np.empty([1,2])
    
    R = d/(2*np.sin(np.pi/K))


    for i in range(K):
        point = np.dot(np.linalg.matrix_power(rotation_matrix,i), np.array([0,R]))
        centres[i,0] = point[0]
        centres[i,1] = point[1]
        x = np.random.normal(point[0],1,N)
        y = np.random.normal(point[1],1,N)
        points = np.concatenate((points,np.array([x,y]).T),axis=0)
    
    points = points[ np.arange(1,N*K+1) ]
    
    plt.plot(centres[:,0], centres[:,1], color='black')
    plt.plot(np.array([centres[0,0], centres[K-1,0]]), np.array([centres[0,1], centres[K-1,1]]), color='black')
    return centres, points



def plot_red_green(K,d,N):
    A, B = calc_clouds(K,d,N)

    dists = np.square(np.sum(np.square(A)[:,np.newaxis,:], axis=2) - 2 * A.dot(B.T) + np.sum(np.square(B), axis=1))
    
    nearest_cloud = np.argmin(dists, axis=0).reshape(K,N)
    nearest_cloud
    
    count_red = 0
    count_green = 0
    
    for (i,j), value in np.ndenumerate(nearest_cloud):
        if nearest_cloud[i,j] !=i:
            plt.scatter(B[N*i+j,0], B[N*i+j,1], color='r', marker = '.')
            count_red +=1
        else:
            plt.scatter(B[N*i+j,0], B[N*i+j,1], color='g', marker = '.')
            count_green += 1
        
    plt.axis('square')
    return count_red, count_green
```

```python

red, green = plot_red_green(3,3.3333,1000)
print(str(np.round(red/green *100,2)) + "%" )
```
