$$X \sim N(2,5) $$
$$Y \sim N(3,1) $$
$$X-Y \sim N(2-3, \sqrt{5^2 + 1^2}) \sim N(-1,\sqrt{26})$$
$$P(X<Y) = P(X-Y<0) =  \Phi \left(\frac{0+1}{\sqrt{26}} \right) \approx 0.57774036626353 $$
