```python
data_x = np.arange(1,11,1)
data_y = np.array([10,10,11,12,18,18,19,26,19,26])
data = np.column_stack((data_x, data_y))
plt.plot(data_x,data_y,'.', color='black')

A = 2
B = -1
C = 5
dists = np.abs(A*data[:,0]+B*data[:,1] + C) / np.sqrt(A**2 + B**2)
dist = np.sum(dists)
print(dist)

x = np.linspace(0,11,100)
y = (-A*x-C) / B
plt.plot(x,y,color='r')
plt.title("dist = " + str(dist))
plt.savefig("zadanie3a.png")
plt.show()
```

```python


def plot_clouds(K,d):

    alfa = 2*np.pi/K
    rotation_matrix = np.array([[np.cos(alfa),np.sin(alfa)],[-np.sin(alfa),np.cos(alfa)]])

    points = np.zeros(2*K).reshape(K,2)

    R = d/(2*np.sin(np.pi/K))


    for i in range(K):
        point = np.dot(np.linalg.matrix_power(rotation_matrix,i), np.array([0,R]))
        points[i,0] = point[0]
        points[i,1] = point[1]
        x = np.random.normal(point[0],1,1000)
        y = np.random.normal(point[1],1,1000)
        plt.plot(x,y,'.')
    

    plt.plot(points[:,0], points[:,1], color='black')
    plt.plot(np.array([points[0,0], points[K-1,0]]), np.array([points[0,1], points[K-1,1]]), color='black')
    plt.axis('square')





```

```python
plot_clouds(7,5)
```
